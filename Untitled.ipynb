#### 1. Price
It appears that listings data with price higher than 25,000 (and including 25,000) isn't accurate. We checked actual listings using the ID and the URL from the master data (combined_df_master). We will just remove these rows.

#### 2. Minimum nights
Given we are focusing on short term rentals, it's safe to limit our data to minimum nights of less than 6 months

#### 3. High number of beds/baths and low price
We have seen a correlation between the nightly price and the number of beds and bathrooms at a listing. It stands to reason that any listing with more than 10 beds or baths would charge a nightly rate of at least $1000. In spot checking these listigns at AirBnB, the results agree with our observation. 

In next steps, we will remove the rows representing the corner cases above. 

```python
condition1 = (combined_df['price'] >= 25000)
condition2 = (combined_df['minimum_nights'] >= 180.00)

# Combine conditions using the AND operator (&)
filtered_df1 = combined_df[~(condition1 + condition2)]

condition3a = (filtered_df1['beds'] > 10)
condition3b = (filtered_df1['price'] < 1000)

filtered_df2 = filtered_df1[~(condition3a & condition3b)]

condition3c = (filtered_df2['bathrooms'] > 10)
condition3d = (filtered_df2['price'] < 1000)

filtered_df3a = filtered_df2[~(condition3c & condition3d)]
filtered_df3a.describe()
```

### Feature Engineering
From correlation matrix, we found that only number of bedrooms and number of bathrooms seemed to have a correlation with the price. 
We want to see if there are other factors that can be gleaned from amenities and description that could help predict the prices.

Some features that typically get higher prices:
1. Pets allowed
3. Free Parking
4. Hot tub
5. Views or beachaccess: Mountain view, Valley view, Ocean view, Beach view or City skyline view, Beachfront
7. Wi-Fi
9. Air conditioning or Heating
10. Indoor fireplace
11. Private entrance
13. Pool 
14. Outdoor area: Patio or balcony or Outdoor dining area
15. Beach access or beach front

We will create these features and conduct analysis to see if any could be relevent to our model.

```python
#create features

#1. pets allowed
filtered_df3=filtered_df3.copy()
filtered_df3['pets'] = np.where(filtered_df3['amenities'].str.contains('Pets allowed'), True, False)

#2. Free Parking
filtered_df3=filtered_df3.copy()
filtered_df3['parking'] = np.where(filtered_df3['amenities'].str.contains('Free parking'), True, False)

#3. Hot tub
filtered_df3=filtered_df3.copy()
filtered_df3['hottub'] = np.where(filtered_df3['amenities'].str.contains('Hot tub|hot tub'), True, False)

#4.views
filtered_df3=filtered_df3.copy()
filtered_df3['views'] = np.where(filtered_df3['amenities'].str.contains('Mountain view|Valley view|Ocean view|Beach view|City skyline view|Beachfront|Oceanfront'), True, False)

#5. Wifi
filtered_df3=filtered_df3.copy()
filtered_df3['Wifi'] = np.where(filtered_df3['amenities'].str.contains('Wifi'), True, False)

#6. Air conditioning and Heating
filtered_df3=filtered_df3.copy()
filtered_df3['ac_heat'] = np.where(filtered_df3['amenities'].str.contains('Air conditioning|Heating'), True, False)

#7. Fireplace
filtered_df3=filtered_df3.copy()
filtered_df3['fireplace'] = np.where(filtered_df3['amenities'].str.contains('Indoor fireplace'), True, False)

#8. Private entrance
filtered_df3=filtered_df3.copy()
filtered_df3['private_entrance'] = np.where(filtered_df3['amenities'].str.contains('Private entrance'), True, False)

#9. Pool
filtered_df3=filtered_df3.copy()
filtered_df3['pool'] = np.where(filtered_df3['amenities'].str.contains('Pool|pool'), True, False)

#10. Outdoor area:Patio or balcony or Outdoor dining area
filtered_df3=filtered_df3.copy()
filtered_df3['outdoor_space'] = np.where(filtered_df3['amenities'].str.contains('Patio|balcony|Outdoor dining area'), True, False)


#11. Beach access or beach front
filtered_df3=filtered_df3.copy()
filtered_df3['beach'] = np.where(filtered_df3['amenities'].str.contains('Beach access|beach front|Beachfront'), True, False)
```
